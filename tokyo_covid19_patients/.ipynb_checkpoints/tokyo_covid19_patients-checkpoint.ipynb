```python
import matplotlib.pyplot as plt
import matplotlib.font_manager
import numpy as np
import pandas as pd

# 定数定義
FOLDER_PATH       = 'tokyo_covid19_patients/'
FONT_PATH         = FOLDER_PATH + 'ipaexm.ttf'
FILE_PATH         = FOLDER_PATH + '130001_tokyo_covid19_patients.csv'
OUTPUT_IMAGE_PATH = FOLDER_PATH + 'pie_chart.png'

# フォント設定(設定しないと日本語が文字化けする)
fp = matplotlib.font_manager.FontProperties(fname=FONT_PATH)

# ファイル読み込み
df = pd.read_csv(FILE_PATH, dtype=object, encoding="utf-8")

# 変数初期化
AGE   = '年代'
label = []
count = np.array([])

# ラベルと件数を設定する
for group in df.groupby(AGE).groups:
    label.append(group)
    count = np.append(count, df.groupby(AGE).size()[group])

# グラフ生成
count, labels, autopct = plt.pie(count, labels=label, autopct="%1.1f%%")

# 日本語化設定
plt.setp(labels, fontproperties=fp)

# グラフを画像出力
plt.savefig(OUTPUT_IMAGE_PATH)

# グラフ表示
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager
import numpy as np
import pandas as pd

# 定数定義
FOLDER_PATH       = 'tokyo_covid19_patients/'
FILE_PATH         = FOLDER_PATH + '130001_tokyo_covid19_patients.csv'
OUTPUT_IMAGE_PATH = FOLDER_PATH + 'bar_graph.png'

# フォント設定(設定しないと日本語が文字化けする)
fp = matplotlib.font_manager.FontProperties(fname=FONT_PATH)

# ファイル読み込み
df = pd.read_csv(FILE_PATH, dtype=object, encoding="utf-8")

# 変数初期化
AGE   = '年代'
label = []
count = np.array([])

# ラベルと件数を設定する
for group in df.groupby(AGE).groups:
    label.append(group)
    count = np.append(count, df.groupby(AGE).size()[group])

# グラフ生成
plt.title('年代別新型コロナ感染数', fontproperties=fp)
plt.bar(label, count)

# グラフを画像出力
plt.savefig(OUTPUT_IMAGE_PATH)

# グラフ表示
plt.show()
```
